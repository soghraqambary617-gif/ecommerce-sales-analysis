```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("online_retail.csv", encoding='ISO-8859-1')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Remove rows without CustomerID
df = df[df['CustomerID'].notna()]

# Remove negative or zero quantities
df = df[df['Quantity'] > 0]

# Remove negative or zero prices
df = df[df['UnitPrice'] > 0]
```

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['Month'] = df['InvoiceDate'].dt.to_period('M')
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
```

```python
df.head()
```

```python
df['Revenue'].sum()
```

```python
monthly_revenue = df.groupby('Month')['Revenue'].sum()
monthly_revenue.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

Revenue Trend:
Revenue increases significantly in the last months of the year (especially NovemberâDecember), likely due to seasonal demand such as holidays.
Top Products:
A small number of products generate a large portion of total revenue, indicating a strong product concentration.


```python
top_products = df.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')
plt.title("Top 10 Products by Revenue")
plt.ylabel("Revenue")
plt.show()
```

```python
top_countries = df.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(10)

top_countries.plot(kind='bar')
plt.title("Top Countries by Revenue")
plt.ylabel("Revenue")
plt.show()
```

Top Country:
Most revenue comes from one main country (likely the UK), showing that the business is highly dependent on a single market.
Some customers contribute significantly more revenue than others, suggesting high-value repeat customers.

```python
day_sales = df.groupby('DayOfWeek')['Revenue'].sum()

day_sales.plot(kind='bar')
plt.title("Revenue by Day of Week")
plt.ylabel("Revenue")
plt.show()
```

Seasonal Revenue Trend:
Revenue peaks significantly in November and December, indicating strong seasonal demand likely driven by holiday shopping behavior.
Geographic Concentration:
The United Kingdom generates the majority of total revenue, showing that the business is highly dependent on a single primary market.
